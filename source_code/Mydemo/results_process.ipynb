```python
# reset columns
# df = df[df['Unnamed: 0'].isin([-100,-50, -30, 0, 30, 60, 90])]


proposed_value = np.array(df['Proposed'])
niqe_value_untrain = np.array(df['NIQE '])
proposed_rank = stats.rankdata(proposed_value)
niqe_rank_untrain = stats.rankdata(niqe_value_untrain)
niqe_value_self_train = np.array(df['NIQE_self_train'])
niqe_rank_self_train = stats.rankdata(-niqe_value_self_train)
niqe_value_train = np.array(df['NIQE_train'])
niqe_rank_train = stats.rankdata(-niqe_value_train) 
brisque_value = np.array(df['BRISQUE'])
brisque_rank = stats.rankdata(-brisque_value)

 
for i in ['f2']:
    # select rows 
    
    value = df[i]
    rank = stats.rankdata(value)
    print(f'{i}: Proposed {stats.spearmanr(proposed_rank, rank)}')
    print(f'{i}: Proposed {stats.pearsonr(proposed_value, value)}')
    
    print(f'{i}: NIQE {stats.spearmanr(niqe_rank_untrain, rank)}')
    print(f'{i}: NIQE {stats.pearsonr(niqe_value_untrain, value)}')
    
    print(f'{i}: NIQE_self_train {stats.spearmanr(niqe_rank_self_train, rank)}')
    print(f'{i}: NIQE_self_train {stats.pearsonr(-niqe_value_self_train, value)}')
    
    print(f'{i}: NIQE_train {stats.spearmanr(niqe_rank_train, rank)}')
    print(f'{i}: NIQE_train {stats.pearsonr(-niqe_value_train, value)}')
    
    print(f'{i}: BRISQUE {stats.spearmanr(brisque_rank, rank)}')
    print(f'{i}: BRISQUE {stats.pearsonr(-brisque_value, value)}')
    
```
