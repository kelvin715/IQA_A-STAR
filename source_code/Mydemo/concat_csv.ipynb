```python
import os
import sys
import pandas as pd

csv_list = []
for i in os.listdir('/Data4/student_zhihan_data/data'):
    if i.endswith('.csv'):
        csv_list.append(i)

csv_list.sort()
```

```python
csv_list
```

```python
pds = [pd.read_csv(os.path.join('/Data4/student_zhihan_data/data', i)) for i in csv_list]
```

```python
img_name_set = set(pds[0].img_name)
img_name_set = img_name_set.intersection(set(pds[2].img_name))
img_name_set
```

```python
for pd in pds:
    img_name_set = img_name_set.intersection(set(pd.img_name))
    
img_name_set = list(img_name_set)
img_name_set.sort()
len(img_name_set)
```

```python
# concat rows

for i in range(len(pds)):
    pds[i] = pds[i][pds[i].img_name.isin(img_name_set)]
```
